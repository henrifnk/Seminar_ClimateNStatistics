```python
dfs = []
for filename in stream_dir:
    df = pd.read_csv(os.path.join('..',path, filename),sep=';',skiprows=10,parse_dates = True,decimal = ',')
    df['measurement_location'] = filename[:8]
    dfs.append(df)
streamflow_data = pd.concat(dfs, ignore_index=True)
```

```python
streamflow_data
```

```python

# Assuming your DataFrame is named df
pivoted_df = pd.pivot_table(streamflow_data, values='Mittelwert', index='Datum', columns='measurement_location')

# Resetting index to make Datum a column again
pivoted_df.reset_index(inplace=True)

# Displaying the pivoted DataFrame
pivoted_df
```

```python
#check for missing values
pivoted_df.isna().sum()
#exclude 1006200 and 15242508 because of too many missing values. the Locations of the 2 stations should alow this
#impute the rest via interpolation of the two adjacent points
```

```python
stream_df = pivoted_df.drop(columns=['15242508','10062000'])
stream_df.interpolate(method='linear', inplace=True)
stream_df.isna().sum()
```

```python
stream_df.to_pickle(os.path.join('..','data','processed','stream_processed.pkl'),compression='zip')
```

```python
def exponential_smoothing(series, alpha):
    result = [series[0]] # First value is same as series
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
    return result
```

```python
plt.plot(stream_df['15247002'].rolling(window=11).mean())
```

```python
# Example usage
# Assuming 'data' is your time series data
alpha = 0.2 # Smoothing parameter (0 < alpha < 1)
smoothed_data = exponential_smoothing(stream_df['15247002'], alpha)

plt.figure(figsize=(20, 14))

plt.plot(stream_df['15247002'])

plt.plot(stream_df['15247002'].rolling(window=3).mean())
plt.plot(smoothed_data)
```
