```python
import pandas as pd
import os
```

```python
path = 'data/raw/streamflow'
stream_dir = os.listdir(os.path.join('..',path))
stream_dir
```

```python
dfs = []
for filename in stream_dir:
    df = pd.read_csv(os.path.join('..',path, filename),sep=';',skiprows=10,parse_dates = True,decimal = ',')
    df['measurement_location'] = filename[:8]
    dfs.append(df)
streamflow_data = pd.concat(dfs, ignore_index=True)
```

```python
streamflow_data.info()
```

```python

# Assuming your DataFrame is named df
pivoted_df = pd.pivot_table(streamflow_data, values='Mittelwert', index='Datum', columns='measurement_location')

# Resetting index to make Datum a column again
pivoted_df.reset_index(inplace=True)

# Displaying the pivoted DataFrame
pivoted_df
```
